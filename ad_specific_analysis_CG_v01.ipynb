```python
%matplotlib inline
import pickle
import pandas as pd
import datetime
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_pickle('data/pres_sorted_data.pickle')
#print(df.dtypes)
df['date_aired'] = df['start_time'].map(lambda x: x.date())
print(df['date_aired'].head())
```

```python
## ad counts by archive ID#
print(df.archive_id.value_counts())
```

```python
df_01 = df.groupby('archive_id')
top_loc_counts = df[df['archive_id'] == 'PolAd_HillaryClinton_DonaldTrump_o35rq']['location'].value_counts()
print(top_loc_counts.head())
top_loc_counts.plot(kind='bar')
#df_01[df_01['archive_id'] == 'PolAd_HillaryClinton_DonaldTrump_o35rq']['start_time'].date().value_counts()
```

```python
## Most aired ad over time
top_date_counts = df[df['archive_id'] == 'PolAd_HillaryClinton_DonaldTrump_o35rq']['date_aired'].value_counts()
print(top_date_counts.head())
top_date_counts.plot()
```
